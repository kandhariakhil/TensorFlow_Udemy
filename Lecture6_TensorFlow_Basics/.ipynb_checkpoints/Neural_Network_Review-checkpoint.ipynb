# Simple Regression Example

```python
import tensorflow as tf
import numpy as np
```

```python
np.random.seed(101)
tf.set_random_seed(101)
```

```python
x_data = np.linspace(0,10,10)+np.random.uniform(-1.5,1.5,10)
```

```python
x_data
```

```python
y_label = np.linspace(0,10,10)+np.random.uniform(-1.5,1.5,10)
```

```python
y_label
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(x_data,y_label,'*')
plt.grid(True)
```

```python
m = tf.Variable(0.5)
b = tf.Variable(0.1)
```

```python
x_data[2]
```

```python
error = 0

for i in range(len(x_data)):
    y_hat = m*x_data[i]+b
    
    error += (y_label[i]-y_hat)**2
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.0001)

train = optimizer.minimize(error)
```

```python
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    
    training_steps = 10000
    
    for i in range(training_steps):
        sess.run(train)
    
    final_slope,final_intercept = sess.run([m,b])
    
```

```python
print(final_intercept)
```

```python
print(final_slope)
```

```python
plt.plot(x_data,y_label,'*',np.linspace(0,11,10),np.linspace(0,11,10)*final_slope+final_intercept)
```

# Regression Example 2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import tensorflow as tf
```

```python
x_data = np.linspace(0,10,100000)+np.random.randn(len(x_data))
```

```python
noise = np.random.randn(len(x_data))
```

```python
noise.shape
```

```python
y_label = (0.5*x_data)+5+noise
```

```python
plt.plot(x_data,y_label)
```

```python
x_df = pd.DataFrame(data=x_data,columns=['X Data'])
y_df = pd.DataFrame(y_label,columns=['Y Label'])
```

```python
y_df.head()
```

```python
my_data = pd.concat([x_df,y_df],axis = 1)
```

```python
my_data.head()
```

```python
m = tf.Variable(0.1)
b = tf.Variable(1.0)
```

```python
batch_size = 10
```

```python
xph = tf.placeholder(tf.float32,shape=batch_size)
yph = tf.placeholder(tf.float32,shape=batch_size)
```

```python
y_hat = m*xph+b
```

```python
error = tf.reduce_sum(tf.square(yph-y_hat))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.0001)
train = optimizer.minimize(error)
```

```python
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    
    sess.run(init)
    
    batches = 100000
    
    for i in range(batches):
    #Pick random 10 data points from x and y 
        rand_ind = np.random.randint(len(x_data),size = batch_size)
        
        #Feed the 10 random data points
        
        feed = {xph:x_data[rand_ind],yph:y_label[rand_ind]}
        
        sess.run(train, feed_dict = feed)
        
    model_m, model_b = sess.run([m,b])
```

```python
model_m
```

```python
model_b
```

```python
y_calculated = model_m*x_data+model_b
```

```python
plt.plot(x_data,y_label,'*',x_data,y_calculated,'r')
```

# Regression Using Estimator API

```python
import numpy as np
```

```python
import tensorflow as tf
```

```python
feat_cols = [tf.feature_column.numeric_column('x',shape=[1])]
```

```python
x_data = np.linspace(0,10,100000)+np.random.randn(100000)
y_label = (0.5*x_data)+5+np.random.randn(100000)
```

```python
estimator = tf.estimator.LinearRegressor(feature_columns=feat_cols)
```

```python
from sklearn.model_selection import train_test_split
```

```python
len(x_data)
```

```python
len(y_label)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, y_label, test_size=0.33, random_state=101)
```

```python
input_func = tf.estimator.inputs.numpy_input_fn({'x':X_train},y_train,batch_size=8,num_epochs=None,shuffle=True)
```

```python
train_input_func = tf.estimator.inputs.numpy_input_fn({'x':X_train},y_train,batch_size=8,num_epochs=1000,shuffle=False)
```

```python
eval_input_func = tf.estimator.inputs.numpy_input_fn({'x':X_test},y_test,batch_size=8,num_epochs=1000,shuffle=False)
```

```python
estimator.train(input_fn=input_func,steps = 1000)
```

```python
train_metrics = estimator.evaluate(input_fn=train_input_func,steps=1000)
```

```python
eval_metrics = estimator.evaluate(input_fn=eval_input_func,steps=1000)
```

```python
print(train_metrics)
```

```python
print(eval_metrics)
```

# Classification Example

```python
import pandas as pd
```

```python
diabetes = pd.read_csv('pima-indians-diabetes.csv')
```

```python
diabetes.head()
```

```python
diabetes.columns
```

```python
cols_to_normalize = ['Number_pregnant', 'Glucose_concentration', 'Blood_pressure', 'Triceps',
       'Insulin', 'BMI', 'Pedigree']
```

```python
diabetes[cols_to_normalize] = diabetes[cols_to_normalize].apply(lambda x: (x-x.min())/(x.max()-x.min()))
```

```python
diabetes.head()
```

```python
import tensorflow as tf
```

```python
num_preg = tf.feature_column.numeric_column('Number_pregnant')
```

```python
plasma_gluc = tf.feature_column.numeric_column('Glucose_concentration')
dias_press = tf.feature_column.numeric_column('Blood_pressure')
tricep = tf.feature_column.numeric_column('Triceps')
insulin = tf.feature_column.numeric_column('Insulin')
bmi = tf.feature_column.numeric_column('BMI')
diabetes_pedigree = tf.feature_column.numeric_column('Pedigree')
age = tf.feature_column.numeric_column('Age')
```

```python
assigned_group = tf.feature_column.categorical_column_with_vocabulary_list('Group',['A','B','C','D'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
diabetes['Age'].hist(bins = 20)
```

```python
age_bucket = tf.feature_column.bucketized_column(age,boundaries=[20,30,40,50,60,70,80])
```

```python
feat_cols = [num_preg,plasma_gluc,dias_press,tricep,insulin,bmi,diabetes_pedigree,assigned_group,age_bucket]
```

```python
#Train-test Split
```

```python
diabetes.head()
```

```python
x_data = diabetes.drop('Class',axis=1)
```

```python
x_data
```

```python
labels = diabetes['Class']
```

```python
labels
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, labels, test_size=0.3, random_state=101)
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size=10,num_epochs=1000,shuffle=True)
```

```python
model = tf.estimator.LinearClassifier(feature_columns=feat_cols,n_classes=2)
```

```python
model.train(input_fn=input_func,steps=1000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test,y=y_test,batch_size=10,num_epochs=1000,shuffle=True)
```

```python
results = model.evaluate(eval_input_func,steps=1000)
```

```python
results
```

```python
pred_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test,batch_size=10,num_epochs=1,shuffle=False)
```

```python
predictions = model.predict(pred_input_func)
```

```python
my_pred = list(predictions)
```

```python
my_pred
```

```python
dnn_model =tf.estimator.DNNClassifier(hidden_units=[10,10,10],feature_columns=feat_cols,n_classes=2)
```

```python
dnn_model.train(input_fn=input_func,steps=1000)
```

```python
#Error for DNN classifier for assigned group
embedded_group_cols = tf.feature_column.embedding_column(assigned_group,dimension=4)
```

```python
feat_cols = [num_preg,plasma_gluc,dias_press,tricep,insulin,bmi,diabetes_pedigree,embedded_group_cols,age_bucket]
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size=10,num_epochs=1000,shuffle=False)
```

```python
dnn_model.train(input_fn=input_func,steps=1000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test,y=y_test,batch_size=10,num_epochs=1000,shuffle=True)
```

```python
dnn_model.evaluate(eval_input_func,steps=1000)
```
